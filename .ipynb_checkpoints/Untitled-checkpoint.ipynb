```python
#Import Library yang dibutuhkan
import pandas as pd
import math
import random
import matplotlib.pyplot as plt
import copy
import numpy
from shapely.geometry import Polygon as pol
```

```python
#Fungsi untuk mengambil titik koordinat yang ada pada suatu objek
def getCoordinate(dataframe):
    x = list(dataframe.lon)
    y = list(dataframe.lat)
    return [x, y]
```

```python
#Fungsi mengambil luas dari dataframe
def getLuas(dataframe):
    return list(dataframe.luas)
```

```python
#Fungsi Faktorisasi
def factor(number):
    x = []
    x.clear()
    for i in range(1, number + 1):
        if number % i == 0:
            x.append(i)
    return x
```

```python
#Fungsi mencari kombinasi panjang dan lebar tiap objek tanah
def creatingCombinationXY(arrLuas):
    combination = []
    combination.clear()
    for i in arrLuas:
        xy = []
        factorXY = factor(i)
        rFactorXY = factorXY[::-1]
        if (len(factorXY) == 4):
            xy.append([factorXY[1], rFactorXY[1]])
            xy.append([factorXY[2], rFactorXY[2]])
        else:
            for j in range(2, len(factorXY) - 2):
                xy.append([factorXY[j], rFactorXY[j]])
        combination.append(xy)
    return combination
```

```python
#Fungsi menciptakan chromosome
def createChromosome(arrCombination):
    chromosome = []
    chromosome.clear()
    for i in arrCombination:
        rng = random.randint(0, len(i)-1)
        rngX = random.uniform(lowerBoundX, upperBoundX)
        rngY = random.uniform(lowerBoundY, upperBoundY)
        chromosome.append([[i[rng][0] * converter, i[rng][1] * converter], rngX, rngY])
    return chromosome
```

```python
#Fungsi mengubah parameter panjang dan lebar menjadi persegi
def createShape(gene):
    x = [0+gene[1], gene[0][0]+gene[1], gene[0][0]+gene[1], 0+gene[1], 0+gene[1]]
    y = [0+gene[2], 0+gene[2], -gene[0][1]+gene[2], -gene[0][1]+gene[2], 0+gene[2]]
    return [x, y]
```

```python
#Fungsi mengambil setiap titik yang terdapat pada suatu objek
def getPoint(shape):
    point = []
    x = shape[0]
    y = shape[1]
    for i in range(len(x)-1):
        point.append((x[i], y[i]))
    return point
```

```python
#Fungsi untuk melakukan pengecekan overlap antar objek tanah
def isOverlapping(shapes, shape):
    goal = len(shapes)
    counter = 0
    pShape = pol(getPoint(shape))
    for i in shapes:
        pOT = pol(getPoint(i))
        if (pShape.intersects(pOT) == False):
            counter += 1
    return counter == goal
```

```python
#Fungsi untuk melakukan pengecekan, apakah objek tanah overlap dengan objek non objek tanah
def isOverlappingWithNonOT(nObjekTanah, shape):
    goal = len(nObjekTanah)
    counter = 0
    pShape = pol(getPoint(shape))
    for i in nObjekTanah:
        pNOT = pol(getPoint(i))
        if (pNOT.intersects(pShape) == False):
            counter += 1
    return counter == goal
```

```python
#Fungsi untuk melakukan pengecekan, apakah objek tanah dipetakan tepat di dalam Main Area
def isInside(mArea, shape):
    pMA = pol(getPoint(mArea))
    pShape = pol(getPoint(shape))
    return pMA.contains(pShape)
```

```python
#Fungsi untuk menghitung nilai fitness dari sebuah chromosome
def fitnessFunction(chromosome):
    shapes = []
    shapes.clear()
    for i in chromosome:
        shape = createShape(i)
        shapes.append(shape)
    fValue = 0
    for j in range(len(shapes)):
        k = j + 1
        shapesOT = shapes[k:]
        if (j == len(shapes)):
            if ((isInside(mArea, shapes[j]) == True) and (isOverlappingWithNonOT(nonOTs, shapes[j]) == True)):
                fValue += 1
        if ((isInside(mArea, shapes[j]) == True) and (isOverlappingWithNonOT(nonOTs, shapes[j]) == True) and (isOverlapping(shapesOT, shapes[j]) == True)):
            fValue += 1
    return fValue
```

```python
#Fungsi menciptakan populasi
def createPopulation(arrCombination, populationSize):
    population = []
    population.clear()
    for i in range(populationSize):
        nChromosome = createChromosome(arrCombination)
        fitness = fitnessFunction(nChromosome)
        population.append({"chromosome": nChromosome, "fitness": fitness})
    return population
```

```python
# Fungsi memilih chromosome terbaik
def selectedChromosome(population):
    sPopulation = sorted(population, key = lambda i: i['fitness'], reverse = True)
    sChromosome1 = copy.deepcopy(sPopulation[0])
    sChromosome2 = copy.deepcopy(sPopulation[1])
    return sChromosome1, sChromosome2
```

```python
#Fungsi untuk melakukan crossover
def crossover(chromosome1, chromosome2):
    divider = int(math.ceil(len(chromosome1["chromosome"]) / 2))
    dividedC1 = [chromosome1["chromosome"][i * divider:(i + 1) * divider] for i in range((len(chromosome1["chromosome"]) + divider - 1) // divider )]
    dividedC2 = [chromosome2["chromosome"][i * divider:(i + 1) * divider] for i in range((len(chromosome2["chromosome"]) + divider - 1) // divider )]
    crossoverC1 = dividedC1[0] + dividedC2[1]
    fitnessC1 = fitnessFunction(crossoverC1)
    crossoverC2 = dividedC2[0] + dividedC1[1]
    fitnessC2 = fitnessFunction(crossoverC2)
    newC1 = {"chromosome": crossoverC1, "fitness": fitnessC1}
    newC2 = {"chromosome": crossoverC2, "fitness": fitnessC2}
    return newC1, newC2
```

```python
#Fungsi untuk memutasi gen yang ada pada chromosome
def mutation(chromosome):
    mutationRate = 0.1
    for i in range(len(chromosome["chromosome"])):
        rngMutation = random.random()
        if (rngMutation <= mutationRate):
            rngIndex = random.randint(0, len(combinationXY[i]) - 1)
            rngX = random.uniform(lowerBoundX, upperBoundX)
            rngY = random.uniform(lowerBoundY, upperBoundY)
            chromosome["chromosome"][i][0] = [combinationXY[i][rngIndex][0] * converter, combinationXY[i][rngIndex][1] * converter]
            chromosome["chromosome"][i][1] = rngX
            chromosome["chromosome"][i][2] = rngY
    fitness = fitnessFunction(chromosome["chromosome"])
    chromosome["fitness"] = fitness
    return chromosome
```

```python
#Fungsi memasukkan chromosome yang terpilih ke dalam populasi
def insertChromosome(chromosome1, chromosome2, population):
    sPopulation = sorted(population, key = lambda i: i['fitness'], reverse = True)
    sPopulation.pop()
    sPopulation.pop()
    sPopulation.append(chromosome1)
    sPopulation.append(chromosome2)
    return sPopulation
```

```python
#Fungsi pemetaan setiap objek tanah ke dalam Main Area
def plotting(chromosome, mArea, nOTs):
    shapes = []
    shapes.clear()
    for i in chromosome["chromosome"]:
        shape = createShape(i)
        shapes.append(shape)
    plt.figure(figsize=(20,10))
    a, b = mArea
    plt.plot(a, b, 'g')
    for nOT in nOTs:
        x, y = nOT
        plt.fill(x, y, 'c')
    for shape in shapes:
        x, y = shape
        plt.plot(x, y)
    return plt.draw()
```
